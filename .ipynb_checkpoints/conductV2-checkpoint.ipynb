## Paso 1:
### Geometría/Malla

- $l_x$: largo en dirección x
- $l_y$: largo en dirección y
- $ncvx$: número de volúmenes de control en dirección x
- $ncvy$: número de volúmenes de control en dirección y

```python
def geometry2d():
    lx = 0.11
    ly = 0.11
    ncvx = 25
    ncvy = 25
    return lx, ly, ncvx, ncvy
```

## Paso 2:
### Inicializar

- Temperatura inicial: $T_0$, (phi[ : ])
- Tiempo de simulación: timeEnd
- Paso de tiempo: $\Delta t$, (dt)

```python
def initialize(phi):
    t0 = 25.
    phi[:] = t0
    timeEnd = 144010.
    dt = 10.
    return phi, timeEnd, dt
```

## Paso 3:
### $\Gamma$
- Calcular $\Gamma$. El resultado escribirlo al lado de "return"

```python
@jit
def gammau(t):
    gam = fp.condf(0.6599, 0.186, 0.1506, 0., 0., 0.0079, 0., t)
    return gam
```

## Paso 4:
### $\lambda$
- Calcular $\lambda$. El resultado escribirlo al lado de "return"

```python
@jit
def lambu(t):
    cp = fp.cpf(0.6599, 0.186, 0.1506, 0., 0., 0.0079, 0., t)
    rho = fp.rhof(0.6599, 0.186, 0.1506, 0., 0., 0.0079, 0., t) 
    rhocp = rho * cp
    return rhocp
```

## Paso 5:
### Selección del solver
- Recomendaciones de solver (sol): 
  - Malla pequeña, utilizar la opción 2
  - Malla grande, utilizar la opción 1
- Tolerancia recomendada (tol): $1\times 10^{-6}$


```python
def solverSettings():
# 1: CG;  2: Dierect; 3: GMRES;  4: LGMRES;  5: MINRES;  6: QMR
    sol = 2
    tol = 1e-6
    return sol, tol
```

## Paso 6
### Términos fuentes, Bordes

Dirichlet

$$ S_C = T_B \frac{k_B}{\delta x_B} A_B \quad \quad S_P = -\frac{k_B}{\delta x_B} A_B $$

Newman

$$ S_C = q_B A_B \, \,  [W] \quad \quad S_P = 0 $$

Robin

$$ S_C = \left(\frac{h_B \frac{k_B}{\delta x_B}}{h_B + \frac{k_B}{\delta x_B}}\right) T_{\infty} A_B \quad \quad S_P = -\left(\frac{h_B \frac{k_B}{\delta x_B}}{h_B +\frac{k_B}{\delta x_B}}\right) A_B $$

- $\Gamma$, $k_B$: gamma[c]
- Distancia del nodo interno al borde $\delta x_B$: ndist[ i ]
- Volumen de la celda $V$: cvol[ c ]
- Área de interface volumen-borde $A_B$: af[ f ]
- b == 1: borde inferior, b == 2: borde derecho, b == 3: borde superior, b == 4: borde izquierdo

```python
@jit
def sourcesBound(b, c, f, i, ndist, cvol, af, gamma, lamb):
    tinf = 121.
    hb = 600.
    gam = gamma[c]
    dx = ndist[i]
    vol = cvol[c]
    area = af[f]
    
    if(b == 1):
        scb = 0.
        spb = 0.
    elif(b == 2):
        scb = ((hb * gam / dx) / (hb + gam / dx)) * tinf * area
        spb = -((hb * gam / dx) / (hb + gam / dx)) * area
    elif(b == 3):
        scb = ((hb * gam / dx) / (hb + gam / dx)) * tinf * area
        spb = -((hb * gam / dx) / (hb + gam / dx)) * area
    elif(b == 4):
        scb = 0.
        spb = 0.
    else:
        scb = 0.
        spb = 0.
    return scb, spb
```

## Paso 7:
### Términos fuentes, transientes

```python
@jit              
def unsteadySources(phi, phi1, phi2, lamb, dt, cvol, c, titer):
    vol = cvol[c]
    #Euler
    ap0 = lamb[c] / dt * vol
    apts = -ap0 
    acts = ap0 * phi1[c]
    return apts, acts
```

## Paso 8:
### Términos fuentes volumétricos

- $V$ volumen: cvol[ c ]

```python
@jit
def sourcesVol(c, vol):
    scv = 0.
    spv = 0.
    return scv, spv
```

## Paso 9
### Monitores 

- tolNl: error máximo permitido por la no-linealidad
- writeContour: cantidad de pasos de tiempo para crear un archivo que genera un contorno de temperatura
- reportBlackW: cantidad de pasos de tiempo para mostrar en pantalla:
  - Tiempo de simulación transcurrido
  - Cantidad de pasos de tiempo transcurridos
  - Iteraciones requeridas por la no-linealidad en ese paso de tiempo
  - Error alcanzado en las iteraciones por la no-linealidad
- reportContour: cantidad de pasos de tiempo para mostrar en pantalla un contorno de temperatura

```python
def controlSet():
    tolNl = 1e-6 
    alertNl = 100
    writeContour = 250
    reportBlackW = 50
    reportContour = 1000
    return tolNl, writeContour, reportBlackW, reportContour, alertNl
```

## Paso 10
### Post-proceso

- name: nombre del archivo que almacena los datos para los contornos
- ndec: número de decimales 
- cmapStyle: colores del mapa de contornos: https://matplotlib.org/examples/color/colormaps_reference.html
- Recomendación: utilizar un nombre simple, no usar espacios y caracteres poco comunes (ñ, Á, #\$()/, etc)

```python
def postproc():
    name = "contour"
    ndec = 2
    cmapStyle = "jet"
    return name, ndec, cmapStyle
```

## Paso 11
### Monitores

- Genera un archivo con columnas que contienen: tiempo de simulación, valores de la temperatura en los nodos monitoreados
- monCoord: matriz que contiene las coordenadas x e y de cada nodo monitoreado. El programa buscará el nodo más cercano a la coordenada entregada por el usuario
- monNames: vector con los nombres de cada nodo monitoreado
- Recomendación: utilizar un nombre simple, no usar espacios y caracteres poco comunes (ñ, Á, #\$()/, etc)
- reportMon: cantidad de pasos de tiempo para guardar en el archivo los valores de la temperatura en cada nodo monitoreado

```python
def monitors(lx, ly):
    monCoord = np.array([[0., 0.],[lx * 0.5, ly * 0.5], [lx, ly]])
    monNames = np.array(["nodo1", "nodo2", "nodo3"])
    reportMon = 25
    return monCoord, monNames, reportMon
```

## Paso 12
### Ejecucar la siguiente celda $\downarrow$

```python
def userInput(ncvx, ncvy, dt, tol, sol, name, monNames, reportMon, reportBlackW, writeContour, reportContour):
    return ncvx, ncvy, dt, tol, sol, name, monNames, reportMon, reportBlackW, writeContour, reportContour

@jit
def calcSourcesB(indCB, cellC, sc, sp, ndist, cvol, af, gamma, lamb):
    for i in range(len(indCB)):
        ind = indCB[i]       
        c1 = cellC[ind,0]
        idb = cellC[ind,2]
        idf = cellC[ind,3]        
        sci, spi = sourcesBound(idb, c1, idf, ind, ndist, cvol, af, gamma, lamb)
        sc[c1] += sci
        sp[c1] += spi
    return sc, sp

@jit
def calcSourcesV(sc, sp, cvol):
    for i in range(len(sc)):
        sci, spi = sourcesVol(i, cvol[i])
        sc[i] += sci
        sp[i] += spi
    return sc, sp

def calcGammaNP(phi, gamma):
    gamma[:] = gammau(phi[:])
    return gamma

@jit
def calcGamma(phi, gamma):
    for i in range(len(phi)):
        gamma[i] = gammau(phi[i])
    return gamma

def calclambNP(phi, lamb):
    lamb[:] = lambu(phi[:])
    return lamb

@jit
def calclamb(phi, lamb):
    for i in range(len(phi)):
        lamb[i] = lambu(phi[i])
    return lamb


@jit      
def unsteadyScSp(phi, phi1, phi2, lamb, dt, cvol, titer, act, apt):
    for i in range(len(phi)):
        apt0, act0 = unsteadySources(phi, phi1, phi2, lamb, dt, cvol, i, titer)
        act[i] = act0
        apt[i] = apt0
    return act, apt

def diffUnsteady():
    ### INPUT PARAMETERS ###
    lx, ly, ncvx, ncvy = geometry2d()
    name, ndec, cmapStyle = postproc()
    monCoord, monNames, reportMon = monitors(lx, ly)
    tolNl, writeContour, reportBlackW, reportContour, alertNl = controlSet()
    ncvt, phi, phi1, phi2, phif, time, dt, tol, nfb, \
    gamma, sc, sp, ap, b, coef, ndist, ndistf, lamb, apt, act = bb.default(ncvx, ncvy)
    sol, tol = solverSettings()

    ### GEOMETRIC/CONECTIVITY PARAMETERS ###
    xv, yv = bb.ugridVertex2D(ncvx, ncvy, lx, ly) 
    faceV = bb.faceVertexC(ncvx, ncvy)
    vertexC, vertexID = bb.vertex(ncvx, ncvy)
    cellC = bb.cellConect(ncvx, ncvy)
    indCI, indCB, indCN, indCII = bb.indexC(ncvx, ncvy, cellC)
    xc, yc = bb.cellCenter2(vertexC, xv, yv, ncvx, ncvy)
    xf, yf = bb.faceCenter(faceV, xv, yv, ncvx, ncvy)
    af = bb.Af(ncvx, ncvy, faceV, xf, yf, xv, yv)
    nf = bb.faceNormal(indCN, cellC, xc, yc)
    cvol = bb.volume(vertexC, xv, yv, ncvx, ncvy)
    bcnS, bcnE, bcnN, bcnW, sources = bb.boundaryIndex(ncvx, ncvy, cellC, indCB)
    ndist, ndistf = bb.nodalDist2(ncvx, ncvy, cellC, xc, yc, xf, yf, ndist, ndistf)
    I = cellC[indCI,0]
    J = cellC[indCI,1]

    nodesM = bb.findNodes(monCoord, xv, yv, xc, yc, vertexC)
    timer = 0.
    titer = 0

    ### INITIALIZATION ###
    phi, timeEnd, dt = initialize(phi)
    phi1[:] = phi[:]
    phi2[:] = phi[:]

    archMonitors = open("monitors.csv", "w")

    while timer < timeEnd:
        
        errNl = 1.
        countIterNl = 0
        
        while errNl >= tolNl:
            
            phi0Nl = phi
            gamma = calcGammaNP(phi, gamma)
            lamb = calclambNP(phi, lamb)
            #Sources
            sc, sp = calcSourcesB(indCB, cellC, sc, sp, ndist, cvol, af, gamma, lamb)
            sc, sp = calcSourcesV(sc, sp, cvol)
            act, apt = unsteadyScSp(phi, phi1, phi2, lamb, dt, cvol, titer, act, apt)

            diffc = bb.diffCoef(ndistf, gamma, indCN, cellC, af)
            ap = bb.calcAp(indCN, cellC, sp, diffc, ap, apt)
            b = bb.calcbNP(sc, act, b)
            vr = bb.val(cellC, indCN, indCII, diffc, ap)
            #Matrix A
            A = bb.matrixAssemble(indCI, vr, cellC, ncvx, ncvy, I, J)
            #solve
            phi = bb.solvers(phi, b, A, tol, sol)
            phif = bb.solverBound2(nfb, cellC, indCB, ndist, sc, sp, phi, gamma, phif)

            vr, ap, sc, sp = bb.resetTime(vr, ap, sc, sp)
            errNl = np.mean((phi0Nl - phi)**2)
            phi0Nl = phi
            countIterNl += 1
            if countIterNl > alertNl:
                print("more than 100 iterations have been executed due to non-linearity")
            
        #Postprocessing
        if titer % writeContour == 0:
            bb.writeFile(cellC, indCB, ncvx, ncvy, xc, yc, xf, yf, phi, phif, name, ndec , timer)
        if titer % reportContour == 0:
            bb.writeFile(cellC, indCB, ncvx, ncvy, xc, yc, xf, yf, phi, phif, name, ndec , timer)
            bb.contour2(lx, ly, ncvx, ncvy, name, timer, ndec, cmapStyle)

        bb.generateMonitors(archMonitors, phi, titer, monNames, nodesM, reportMon, timer)

        #update phi
        phi2[:] = phi1[:]
        phi1[:] = phi[:]

        #next time step
        timer += dt
        titer += 1

        if titer % reportBlackW == 0:
            print("time = " + str(timer) + "  iter_t = " +  str(titer) + "  nl_iter = " \
                  + str(countIterNl) + "  errNl = " + str(errNl))
                 
    archMonitors.close()
    print("calculation complete")
```
